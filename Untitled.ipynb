```python
import random
import pandas as pd
import folium
from math import radians, sin, cos, sqrt, atan2
import matplotlib.pyplot as plt

# Функция для вычисления расстояния между двумя точками (города) по формуле Хаверсина
def haversine(lat1, lon1, lat2, lon2):
    R = 6371  # Радиус Земли в километрах
    dlat = radians(lat2 - lat1)
    dlon = radians(lon2 - lon1)
    a = sin(dlat / 2)**2 + cos(radians(lat1)) * cos(radians(lat2)) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))
    return R * c

# Чтение данных городов из CSV
cities_df = pd.read_csv('kz.csv')
cities = cities_df[['city', 'lat', 'lng']].to_dict(orient='records')

# Добавление стартового города Qulsary
qulsary = {'city': 'Qulsary', 'lat': 46.9833, 'lng': 54.0167}
cities.insert(0, qulsary)

# Алгоритм ближайшего соседа
def nearest_neighbor(cities):
    route = [0]
    unvisited = set(range(1, len(cities)))
    while unvisited:
        last_visited = route[-1]
        closest_city = min(unvisited, key=lambda city: haversine(
            cities[last_visited]['lat'], cities[last_visited]['lng'],
            cities[city]['lat'], cities[city]['lng']))
        route.append(closest_city)
        unvisited.remove(closest_city)
    return route

# Генетический алгоритм
def genetic_algorithm(cities, population_size=100, generations=500, mutation_rate=0.1):
    def create_individual():
        individual = list(range(1, len(cities)))
        random.shuffle(individual)
        return [0] + individual

    def fitness(individual):
        return sum(haversine(
            cities[individual[i]]['lat'], cities[individual[i]]['lng'],
            cities[individual[i + 1]]['lat'], cities[individual[i + 1]]['lng']
        ) for i in range(len(individual) - 1))

    def mutate(individual):
        if random.random() < mutation_rate:
            idx1, idx2 = random.sample(range(1, len(individual)), 2)
            individual[idx1], individual[idx2] = individual[idx2], individual[idx1]

    def crossover(parent1, parent2):
        start, end = sorted(random.sample(range(1, len(parent1)), 2))
        child = [-1] * len(parent1)
        child[start:end] = parent1[start:end]
        pointer = 0
        for gene in parent2:
            if gene not in child:
                while child[pointer] != -1:
                    pointer += 1
                child[pointer] = gene
        return child

    population = [create_individual() for _ in range(population_size)]
    for _ in range(generations):
        population = sorted(population, key=fitness)
        next_generation = population[:10]
        while len(next_generation) < population_size:
            parents = random.sample(population[:50], 2)
            child = crossover(parents[0], parents[1])
            mutate(child)
            next_generation.append(child)
        population = next_generation
    return population[0]

# 2-opt алгоритм
def two_opt(cities):
    def calculate_total_distance(route):
        return sum(haversine(
            cities[route[i]]['lat'], cities[route[i]]['lng'],
            cities[route[i + 1]]['lat'], cities[route[i + 1]]['lng']
        ) for i in range(len(route) - 1)) + haversine(
            cities[route[-1]]['lat'], cities[route[-1]]['lng'],
            cities[route[0]]['lat'], cities[route[0]]['lng'])

    route = list(range(len(cities)))
    improved = True
    while improved:
        improved = False
        for i in range(1, len(route) - 2):
            for j in range(i + 1, len(route)):
                if j - i == 1: continue
                new_route = route[:i] + route[i:j][::-1] + route[j:]
                if calculate_total_distance(new_route) < calculate_total_distance(route):
                    route = new_route
                    improved = True
    return route

# Визуализация маршрута на карте
def visualize_route_html(route, cities, algorithm_name):
    m = folium.Map(location=[cities[0]['lat'], cities[0]['lng']], zoom_start=5)
    for city in cities:
        folium.Marker([city['lat'], city['lng']], popup=city['city']).add_to(m)
    for i in range(len(route) - 1):
        city1 = cities[route[i]]
        city2 = cities[route[i + 1]]
        folium.PolyLine([(city1['lat'], city1['lng']), (city2['lat'], city2['lng'])], color="blue", weight=2.5).add_to(m)
    city1 = cities[route[-1]]
    city2 = cities[route[0]]
    folium.PolyLine([(city1['lat'], city1['lng']), (city2['lat'], city2['lng'])], color="blue", weight=2.5).add_to(m)
    m.save(f"{algorithm_name}_route.html")

# Визуализация сравнения маршрутов
routes = {
    "Nearest Neighbor": nearest_neighbor(cities),
    "Genetic Algorithm": genetic_algorithm(cities),
    "2-opt": two_opt(cities)
}

# Сравнение расстояний
distances = {}
for name, route in routes.items():
    distance = sum(haversine(
        cities[route[i]]['lat'], cities[route[i]]['lng'],
        cities[route[i + 1]]['lat'], cities[route[i + 1]]['lng']
    ) for i in range(len(route) - 1)) + haversine(
        cities[route[-1]]['lat'], cities[route[-1]]['lng'],
        cities[route[0]]['lat'], cities[route[0]]['lng'])
    distances[name] = distance
    visualize_route_html(route, cities, name)


plt.bar(distances.keys(), distances.values(), color=['blue', 'green', 'red'])
plt.ylabel('Distance (km)')
plt.title('Comparison of TSP Algorithms')
plt.show()
```

```python
# Функция для создания анимации графика маршрута
def create_route_gif(route, cities, algorithm_name, output_folder='output_gifs'):
    images = []
    
    # Для каждого шага маршрута создаем график
    for i in range(1, len(route) + 1):
        # Параметры маршрута до текущего шага
        route_cities = [cities[i] for i in route[:i]]  # Только те города, которые включены в текущий маршрут
        
        # Строим график
        fig, ax = plt.subplots(figsize=(10, 8))
        latitudes = [city['lat'] for city in route_cities]
        longitudes = [city['lng'] for city in route_cities]
        
        ax.scatter(longitudes, latitudes, color='red')

        for j, city in enumerate(route_cities):
            ax.text(longitudes[j] + 0.2, latitudes[j], city['city'], fontsize=9)

        # Соединяем города маршрута
        for j in range(len(route_cities) - 1):
            city1 = route_cities[j]
            city2 = route_cities[j + 1]
            ax.plot([city1['lng'], city2['lng']], [city1['lat'], city2['lat']], color='blue')

        ax.set_title(f'{algorithm_name} Algorithm - Step {i}')
        ax.set_xlabel('Longitude')
        ax.set_ylabel('Latitude')
        ax.grid(True)

        # Сохраняем изображение в список
        image_path = os.path.join(output_folder, f'{algorithm_name}_step_{i}.png')
        plt.savefig(image_path)
        images.append(image_path)
        plt.close(fig)
    
    # Создаем GIF
    gif_path = os.path.join(output_folder, f'{algorithm_name}_animation.gif')
    with imageio.get_writer(gif_path, mode='I', duration=0.5) as writer:
        for image in images:
            writer.append_data(imageio.imread(image))
            os.remove(image)  # Удаляем временное изображение
    print(f'GIF сохранен как {gif_path}')

# Пример вызова для трех алгоритмов
def generate_all_gifs(cities, algorithm_results):
    output_folder = 'output_gifs'
    os.makedirs(output_folder, exist_ok=True)
    
    for algorithm_name, route in algorithm_results.items():
        create_route_gif(route, cities, algorithm_name, output_folder)

# Генерация маршрутов с использованием алгоритмов
nn_route = nearest_neighbor(cities)
ga_route = genetic_algorithm(cities)
two_opt_route = two_opt(cities)

# Пример алгоритмов с их результатами маршрутов
algorithm_results = {
    'NearestNeighbor': nn_route,
    'GeneticAlgorithm': ga_route,
    'TwoOpt': two_opt_route,
}

# Генерация GIF для всех алгоритмов
generate_all_gifs(cities, algorithm_results)
```
